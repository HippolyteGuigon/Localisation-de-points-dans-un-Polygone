```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from shapely.geometry.polygon import LinearRing, Polygon
from matplotlib import pyplot
from shapely.geometry.polygon import LinearRing, Polygon
from Intersection import *
```

```python
#Entrer une liste de tuples avec chacune des deux coordonnées

def polygon(point, test_point):
    
    counter = 0
    poly = Polygon(point)
    x,y = poly.exterior.xy
    
    ring = LinearRing(point)
    
    x, y = ring.xy
    u = max(x)
    v = max(y)
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)
    ax.plot(x, y, color='#6699cc', alpha=0.7,
    linewidth=3, solid_capstyle='round', zorder=2)
    ax.scatter(test_point[0], test_point[1], color='red', alpha=0.7)
    ax.plot((test_point[0], max(x)+1), (test_point[1], max(y)+1), '-', color="green")
    ax.set_title('Polygône', fontsize=15)

    for i in range(len(point)-1):
        segment_1 = point[i]
        segment_2 = point[i+1]
        counter += 1*intersects((segment_1, segment_2), ((test_point[0],test_point[1] ), (max(x)+1, max(y)+1)))
    counter += intersects((point[0], point[len(point)-1]), ((test_point[0], max(x)+1), (test_point[1], max(y)+1)))
    
    if counter%2 != 0:
        print("Le point est à l'intérieur du polygône")
    else:
        
        print("Le point est à l'extérieur du polygône")
```

```python
polygon([(0, 0), (0, 1), (1, 2), (5, 3)], (1, 1.4))
```

```python
import random
import time

for k in range(3, 10):
    time1 = time.time()
    points = [(random.randint(1, 10), random.randint(1, 12)) for j in range(k)]
    test_point = (random.randint(1, 10), random.randint(1, 12))
    polygon(points, test_point)
    time2 = time.time()
    print("L'opération a durée {} secondes".format(time2-time1))
        
```
